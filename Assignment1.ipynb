```python
"""Q2. Compute the strike rate (runs / balls faced × 100) of top batsmen using ufuncs.
Compare performance in the powerplay (overs 1–6) versus death overs (overs 17–20)."""
```

```python
runs = pd.Series(deliveries.groupby('batter')['total_run'].sum())
balls = pd.Series(deliveries.groupby('batter')['ballnumber'].count())
```

```python
strikeRate = (runs / balls) * 100
```

```python
topBatsmen = runs[runs > 5000].index
topBatsmen
```

```python
sr_top = strikeRate.loc[topBatsmen].sort_values(ascending=False)

sr_top
```

```python
powerplay = deliveries[deliveries['overs'] <= 6]
deathovers = deliveries[deliveries['overs'] > 6]
```

```python
pp_runs = powerplay.groupby('batter')['total_run'].sum()
pp_balls = powerplay.groupby('batter')['ballnumber'].size()

pp_sr = pp_runs / pp_balls * 100
```

```python
do_runs = deathovers.groupby('batter')['total_run'].sum()
do_balls = deathovers.groupby('batter')['ballnumber'].size()

do_sr = do_runs / do_balls * 100
```

```python
comparision = pd.DataFrame({
    'Power Play' : pp_sr,
    'Death over' : do_sr
}).loc[topBatsmen].sort_values(by='Death over', ascending=False)
comparision
```

```python
"""Q4. Create a 'Player Impact Score' combining runs, fours, sixes, and wickets using normalized ufunc weights.
Rank all players and construct an impact leaderboard."""
```

```python
deliveries.fillna(0)
runs = deliveries.groupby('batter')['batsman_run'].sum()
fours = deliveries[deliveries['batsman_run'] == 4].groupby(['batter']).size()
sixes = deliveries[deliveries['batsman_run'] == 6].groupby(['batter']).size()
wickets = deliveries[deliveries['isWicketDelivery'] == 1].groupby(['batter']).size()
```

```python
impact = pd.concat([runs, fours, sixes, wickets], axis=1).fillna(0)
impact.columns=['runs', 'fours',  'sixes', 'wickets']
impact
```

```python
import numpy as np
norm = impact / impact.max()
impact['impact_score'] = np.add.reduce([
    norm['runs'] * 0.4,
    norm['fours'] * 0.2,
    norm['sixes'] * 0.2,
    norm['wickets'] * 0.2
])
impact.sort_values(by='impact_score', ascending=False)
impact.head()
```

```python
"""Q5. Examine missing values in dismissal-related columns. 
Determine the effect of replacing NaN with 'not out' versus dropping those rows on overall bowling statistics."""
```

```python
deliveries[['player_out', 'kind']].isnull().sum()
```

```python
deliveries[['player_out', 'kind']] = deliveries[['player_out', 'kind']].fillna('not out')
deliveries[['player_out', 'kind']].isnull().sum()
```

```python
df_replace = deliveries
wickets_replace = df_replace.groupby('bowler')['isWicketDelivery'].sum()
total_wickets_replace = wickets_replace.sum()

df_drop = df_replace.dropna(subset=['player_out', 'kind'])

wickets_drop = df_drop.groupby('bowler')['isWicketDelivery'].sum()
total_wickets_drop = wickets_drop.sum()

print("Total wickets (NaN replaced):", total_wickets_replace)
print("Total wickets (NaN dropped):", total_wickets_drop)
```

```python
"""Q6. Using boolean indexing, filter deliveries where extras are greater than zero. 
Examine which teams concede the most extras and in which match phases they occur."""
```

```python
df = deliveries.merge(matches, left_on='ID', right_on='ID')

df['bowling_team'] = df.apply(
    lambda col: col['Team1'] if col['BattingTeam'] == col['Team2'] else col['Team2'], axis=1
)

extras = df[df['extras_run'] > 0]


extras['phase'] = extras['overs'].apply(
    lambda x: 'Powerplay' if x<=6 else ('Middle' if x<=15 else 'Death')
)

extras[['BattingTeam', 'extras_run']]
```

```python
mostExtras = extras.groupby('bowling_team')['extras_run'].sum().sort_values(ascending=False)
mostExtras
```

```python
extras.groupby('phase')['extras_run'].sum()
```

```python
"""Q7. Design a match-phase classifier using apply() that labels each delivery as 'Powerplay', 'Middle', or 'Death' overs. 
Build a team-wise phase performance matrix using pivot_table."""
```

```python
import pandas as pd

deliveries['phase'] = deliveries['overs'].apply(
    lambda x: 'Powerplay' if x<=6 else ('Middle' if x<=15 else 'Death')
)

result_matrix = pd.pivot_table(
    deliveries,
    values='total_run',
    index='BattingTeam',
    columns='phase',
    aggfunc='mean'
)

result_matrix
```

```python
"""Q8. Examine how index alignment behaves when merging a bowler-wise wickets Series with a bowler- wise runs-conceded Series. 
Compute economy rate using aligned index arithmetic."""
```

```python
import pandas as pd

wickets = deliveries.groupby('bowler')['isWicketDelivery'].sum()

runs = deliveries.groupby('bowler')['total_run'].sum()

balls = deliveries.groupby('bowler').size()

economy = (runs / balls) * 6

result = pd.concat([wickets, runs, balls, economy], axis=1)
result.columns = ['wickets', 'concededRuns', 'balls', 'economy']

result.head()
```

```python
"""Q9. Using hierarchical indexing on season and team, examine total runs, wickets, and extras per season per team. 
Use unstack() to build a season comparison matrix."""
```

```python
import pandas as pd

df = deliveries.merge(matches[['ID','Season']], on='ID')

stats = df.groupby(['Season','BattingTeam']).agg({
    'total_run': 'sum',
    'isWicketDelivery': 'sum',
    'extras_run': 'sum'
})

matrix = stats.unstack(level=0)

matrix
```
